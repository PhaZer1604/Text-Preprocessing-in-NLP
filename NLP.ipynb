```python
# import dataset
df = pd.read_csv('/content/drive/MyDrive/robert_frost_collection.csv')
df = df.drop(labels=["Collection", "Year of Publication"], axis=1)
df = df.drop(labels=0 , axis=0)
df
```

```python
df['Content'][1]
```

```python
# lowercase data
df['Content'] = df['Content'].str.lower()
df.head()
```

```python
# remove \n from data
import re
df['Content'] = df['Content'].str.replace(r"\n", " ")
df.head()
```

```python
# remove extra whitespaces
df['Content'] = df['Content'].str.strip()
df.head()
```

```python
# remove punctuations 
import string
string.punctuation
def remove_puctuation(text):
    for i in string.punctuation:
        if i in text:
            text = text.replace(i,'')
            
    return text
df['Content'] = df['Content'].apply(remove_puctuation)
df['Content'].head()
```

```python
# spelling correction
from textblob import TextBlob

def spelling_correction(text):
    return TextBlob(text).correct().string
  
df['Content'] = df['Content'].apply(spelling_correction)
df
```

```python
# word tokenize and setence tokenize
import nltk
nltk.download('punkt')

from nltk import word_tokenize,sent_tokenize

df['sentence'] = df['Content'].apply(sent_tokenize)
df['tokenized_content'] = df['Content'].apply(word_tokenize)
df
```

```python
# space tokenize
df['space_token'] = df['Content'].str.split()
df.head()
```

```python
# remove stopwords
nltk.download("stopwords")
from nltk.corpus import stopwords

def remove_stopwords(text):
    L = []
    for word in text:
        if word not in stopwords.words('english'):
            L.append(word)     
    return L

stopwords.words('english')
```

```python
df['tokenized_content'] = df['tokenized_content'].apply(remove_stopwords)
df.head()
```

```python
# Stemming 
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem_words(text):
    return [ps.stem(word) for word in text]

df['stemmed_words'] = df['tokenized_content'].apply(stem_words)
df
```

```python
# Lemmatization
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemma_words(text):
    return [lemmatizer.lemmatize(word) for word in text]

df['lemmatized'] = df['tokenized_content'].apply(lemma_words)
df
```

```python
# spelling checker using Minimum Edit Distance
import itertools
df2 = pd.read_csv('/content/drive/MyDrive/smart home commands.csv')
sentences_df = df2[['Sentence']]
sentences_df.head()
```

```python
def get_plain_vocabluary():
    sentencess = [word_tokenize(sentence['Sentence']) for index, sentence in sentences_df.iterrows()]
    mergesentences = list(itertools.chain.from_iterable(sentencess))
    plainvocabulary = list(set(mergesentences))
    return plainvocabulary
```

```python
def levenshtein_distance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for i2, c2 in enumerate(s2):
        distances_ = [i2+1]
        for i1, c1 in enumerate(s1):
            if c1 == c2:
                distances_.append(distances[i1])
            else:
                distances_.append(1 + min((distances[i1], distances[i1 + 1], distances_[-1])))
        distances = distances_
    return distances[-1]
```

```python
def spelling_correction(sentence):
    splittedsentence = word_tokenize(sentence)
    vocwords = list(itertools.chain.from_iterable([get_plain_vocabluary()]))
    for i,word in enumerate(splittedsentence):
        if (word not in vocwords and not word.isdigit()):
            levdistances = []
            for vocword in vocwords:
                levdistances.append(levenshtein_distance(word,vocword))
            splittedsentence[i] = vocwords[levdistances.index(min(levdistances))]
        else:
            splittedsentence[i] = word
    return ' '.join(splittedsentence)
```
